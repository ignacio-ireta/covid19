```python
import pandas as pd 
```

```python
peru = pd.read_csv(r"PATH", sep=";")  # Because for some reason ";" in a csv is better
```

```python
peru['FECHA_CORTE'] = pd.to_datetime(peru['FECHA_CORTE'], format='%Y%m%d').dt.strftime('%Y-%m-%d')  # Reformat their dates
```

```python
peru['FECHA_RESULTADO'] = pd.to_datetime(peru['FECHA_CORTE'], format='mixed').dt.strftime('%Y-%m-%d')
```

```python
peru.to_csv(r"PATH",sep=',')  # Save with generic corrections
```
